```python
# import libraries 

from bs4 import BeautifulSoup
import requests
import time
import datetime

import smtplib

#Connect to website
URL = 'https://lubimyczytac.pl/ksiazka/20948/dzieje'

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36"}

page = requests.get(URL, headers=headers)

soup1 = BeautifulSoup(page.content ,'html.parser')
print(soup1)

soup2 = BeautifulSoup(soup1.prettify() ,'html.parser')
print(soup2)
```

```python
title = soup2.find(id = "book-info").get_text()
author = soup2.find("a", class_="link-name d-inline-block").get_text()

title = title.strip()
author = author.strip()
print(title)
print(author)
```

```python
import datetime
today = datetime.date.today()

print(today)
```

```python
import csv

header = ['title', 'author', 'date']
data = [title, author, today]

with open('LubimyczytacWebScraper.csv', 'w',newline = '', encoding = 'UTF8') as f:
  writer = csv.writer(f)
  writer.writerow(header)

  writer.writerow(data)
```

```python
def check_price():
  URL = 'https://lubimyczytac.pl/ksiazka/20948/dzieje'

  headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36"}

  page = requests.get(URL, headers=headers)

  soup1 = BeautifulSoup(page.content ,'html.parser')

  soup2 = BeautifulSoup(soup1.prettify() ,'html.parser')

  title = soup2.find(id = "book-info").get_text()
  author = soup2.find("a", class_="link-name d-inline-block").get_text()

  title = title.strip()
  author = author.strip()

  import datetime
  today = datetime.date.today()

  import csv

  header = ['title', 'author', 'date']
  data = [title, author, today]

  with open('LubimyczytacWebScraper.csv', 'w',newline = '', encoding = 'UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(data)
```
